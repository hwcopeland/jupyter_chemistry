Introduction
--------

  Chromium and Cobalt are both transition metals that see most common use in the manufacture of stainless and high-speed steel respectively. Both cobalt and Chromium have stable, water soluable nitrate salts. A realativly simple, non-destructive approach to quantifying $Cr^{3+}$ and $Co^{2+}$ ions in aqueous solution is UV-VIS Absorbance Spectroscopy. For this expiriment we have selected a VIS spectrum analysis of aqueous solution at 510nm to 576nm. First a series of analytical calibration solutions will be produced. The calibration solutions will be then used to fit a linear system of functions. The resulting functions will give the relative concentrations of $Cr^{3+}$ and $Co^{2+}$ ions from the sum of thier absorbances.

Data Ingress and Processing
--------

```python
import re

def parse_filename(filename):
    """Extract ion and molarity from filenames like 'cr3+0.05M.asc'."""
    match = re.match(r"([a-zA-Z0-9\+\-]+)([\d\.]+)M\.asc", filename)
    if match:
        ion, molarity = match.groups()
        return ion[:-1], float(molarity)
    return None, None

def process_folder(folder_path):
    ion_data = {}

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".asc"):
                ion, molarity = parse_filename(file)
                if ion and molarity is not None:
                    file_path = os.path.join(root, file)
                    print(f"Processing: {file_path} (Ion: {ion}, M: {molarity}M)")

                    try:
                        with open(file_path, 'r') as f:
                            lines = f.readlines()

                        wavelengths = []
                        absorbances = []

                        for line in lines:
                            if line.strip() and "\t" in line:
                                parts = line.strip().split("\t")
                                if len(parts) == 2:
                                    try:
                                        wavelength = float(parts[0])
                                        absorbance = float(parts[1])
                                        wavelengths.append(wavelength)
                                        absorbances.append(absorbance)
                                    except ValueError:
                                        continue

                        # Create a DataFrame for the ion if it doesn't exist
                        if ion not in ion_data:
                            ion_data[ion] = pd.DataFrame({'Wavelength (λ)': wavelengths}).set_index('Wavelength (λ)')

                        # Add concentration column
                        ion_data[ion][molarity] = absorbances

                    except Exception as e:
                        print(f"Error processing {file_path}: {e}")

    # Sort concentrations (columns) for each ion
    for ion, df in ion_data.items():
        ion_data[ion] = df[sorted(df.columns)]

    return ion_data

# Example usage (replace with your folder path)
folder_path = "/content/UV-VIS spectral data"  # Replace with your folder path
ion_data = process_folder(folder_path)

# Display each ion's DataFrame
for ion, df in ion_data.items():
    print(f"\nDataFrame for {ion}:")
    print(df)
```

For a system whose absorbances are additive

$$ A_i = \sum_{j=1}k_{ij}C_j $$

for the $ i^{th} $ wavelength.

$$A_i = k_{(i_1,Co^{2+})}C_{(Co^{2+})} + k_{(i_2,Cr^{3+})}C_{(Cr^{3+})}$$

For a given species and the $l^{th}$ concentration and $i^{th}$ wavelength...

$$k_i = \frac{\sum_{l=1}\frac{A_{i,l}}{C_l}}{l_{max}}$$
